## TASK 1: import the dataset

```python
share_url = "https://drive.google.com/file/d/1U_CSIE9CxV4M8XbWoiVANFsbekm6CXmg/view?usp=sharing"
download_url="https://drive.google.com/uc?export=download&id="+share_url.split("/")[-2]
print(download_url)

data = pd.read_csv(download_url)

data.head()
```

```python
import pandas as pd
# 1. How many men and women?
data['sex'].value_counts()

print("Men:", (data['sex'] == 'Male').sum())
print("Women:", (data['sex'] == 'Female').sum())
```

```python
# 2. What is the average age of women?
women_avg_age = data[data['sex'] == 'Female']['age'].mean()
print("Average ago of women:", women_avg_age)
```

```python
# 3. What is the percentage of German citizens?
german_percentage = (data['native-country'] == 'Germany').mean() * 100
print("German citizens:", german_percentage, "%")
```

```python
#4-5. What are the mean and standard deviation of age for those who earn more than 50K per year (salary feature) and those who earn less than 50K per year?
age_stats = data.groupby('income')['age'].agg(['mean', 'std'])
print(age_stats)

print(" >50K:", age_stats.loc['>50K', 'mean'], 
      "std:", age_stats.loc['>50K', 'std'])
print("<=50K:", age_stats.loc['<=50K', 'mean'], 
      "std:", age_stats.loc['<=50K', 'std'])
```

```python
#6. Is it true that people who earn more than 50K have at least high school education? (education – Bachelors, Prof-school, Assoc-acdm, Assoc-voc, Masters or Doctorate feature)

high_education = ['Bachelors', 'Prof-school', 'Assoc-acdm', 'Assoc-voc', 'Masters', 'Doctorate']
high_earners = data[data['income'] == '>50K']
print("Is it true that people who earn more than 50K have at least high school education?", high_earners['education'].isin(high_education).all())
```

```python
#7. Display age statistics for each race (race feature) and each gender (sex feature). Use groupby() and describe(). Find the maximum age of men of Amer-Indian-Eskimo race.
age_by_race_sex = data.groupby(['race', 'sex'])['age'].describe()
print(age_by_race_sex)

max_age_amer_indian_male = data[(data['race'] == 'Amer-Indian-Eskimo') & (data['sex'] == 'Male')]['age'].max()
print("MAX age of men of Amer-Indian-Eskimo:", max_age_amer_indian_male)
```

```python
#8. Among whom is the proportion of those who earn a lot (>50K) greater: married or single men (marital-status feature)? Consider as married those who have a marital-status starting with Married (Married-civ-spouse, Married-spouse-absent or Married-AF-spouse), the rest are considered bachelors.

men = data[data['sex'] == 'Male'].copy()
men['is_married'] = men['marital-status'].str.startswith('Married')

prop_married = men[men['is_married']]['income'].eq('>50K').mean()
prop_single = men[~men['is_married']]['income'].eq('>50K').mean()

print(">50K married:", prop_married)
print(">50K single:", prop_single)
```

```python
#9. What is the maximum number of hours a person works per week (hours-per-week feature)? How many people work such a number of hours, and what is the percentage of those who earn a lot (>50K) among them?

max_h = data['hours-per-week'].max()
people_max = data[data['hours-per-week'] == max_h]
count_max = len(people_max)
percent_high = (people_max['income'] == '>50K').mean() * 100

print("MAX hours/week:", max_h)          
print("People work such a number of hours:", count_max)                      
print("% of those who earn >50K among them:", percent_high)     
```

```python
#10. Count the average time of work (hours-per-week) for those who earn a little and a lot (salary) for each country (native-country). What will these be for Japan?

japan_h = data[data['native-country'] == 'Japan'].groupby('income')['hours-per-week'].mean()

print("Japan <=50K:", japan_h['<=50K'])      
print("Japan >50K :", japan_h['>50K'])       
```
