## Task 3: Read titanic dataset from GitHub repository (CSV file)

```python
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic = pd.read_csv(url)
titanic
```

## 5.1 Select passengers who survived

```python
survived = titanic[titanic['Survived'] == 1]
survived
```

## 5.2. Select passengers who are male and survived

```python
male_survived = titanic[(titanic['Sex'] == 'male') & (titanic['Survived'] == 1)]
male_survived
```

## 5.3. Select passengers who are in Pclass 1 and are older than 30

```python
passengers_set = titanic[(titanic['Pclass'] == 1) & (titanic['Age'] >30)]
passengers_set
```

## 5.4. Select passengers who embarked from Cherbourg ('C') or Queenstown ('Q')

```python
embarked = titanic[titanic['Embarked'].isin(['C', 'Q'])]
embarked
```

## Task 6: (from my head) Use pandas to perform SQL-like operations on the Titanic dataset

```python
# Select specific columns (Name, Age, Survived)
selected_cols = titanic[['Name', 'Age', 'Survived']]
selected_cols
```

```python
# Where: Filter passengers who survived and paid fare > 50
sql_where = titanic[(titanic['Survived'] == 1) & (titanic['Fare'] > 50)]
sql_where
```

```python
# limit: Show top 10 most expensive tickets
sql_limit = titanic.nlargest(10, 'Fare')[['Name', 'Fare', 'Pclass']]
sql_limit
```
