```python
import pandas as pd
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
url = "/home/iren/Downloads/input_image.jpg"
img =io.imread(url)
img
```

```python
plt.imshow(img)
```

```python
# Print the size of the image array
img.shape
```

```python
# Print the image type 
img.dtype
type(img)
```

```python
# Find the max and min pixel values
max_val = np.max(img)
min_val = np.min(img)
print(max_val)
print(min_val)
```

```python
#Padding
padded_image = np.pad(img, pad_width=((20, 20), (20, 20), (0, 0)), mode='constant', constant_values=0)
```

```python
plt.imshow(padded_image)
#np.pad() adds a border to a NumPy array.
```

```python
#Apply np.where, image thresholding
thr_value = 128
bi_img = np.where(img > 0, 255, 0).astype(np.uint8)
plt.imshow(bi_img)
```

```python
#Aplly conv2d, image filtering with kernels
from skimage.filters import sobel, gaussian
```

```python
blurred_image = gaussian(img, sigma=1)
plt.imshow(blurred_image) 
```
