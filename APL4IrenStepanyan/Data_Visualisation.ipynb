<h3>Task 4:  Random Noise Generation and visualisation with different color maps</h3>

use histogram and scatter and plot all three

```python
# Generate random noise data
np.random.seed(42)
noise = np.random.randn(500, 500)

# Plot noise with different color maps
plt.figure(figsize=(15, 5))

# colormap 1: grayscale
plt.subplot(1, 3, 1)
plt.imshow(noise, cmap='gray')
plt.title("Random Noise - Gray")
plt.colorbar()

# colormap 2: plasma
plt.subplot(1, 3, 3)
plt.imshow(noise, cmap='plasma')
plt.title("Random Noise - Plasma")
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
# Generate 1D random noise for histogram and scatter
np.random.seed(42)
noise_1d = np.random.randn(10000)

# Histogram
plt.figure(figsize=(8, 5))
plt.hist(noise_1d, bins=50, color='pink', edgecolor='black')
plt.title("Histogram of 1D Gaussian Noise")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(True, alpha=0.3)
plt.show()

# Scatter plot (index vs value)
indices = np.arange(len(noise_1d))
plt.figure(figsize=(10, 5))
plt.scatter(indices[:1000], noise_1d[:1000], c=noise_1d[:1000], cmap='coolwarm', s=10)
plt.title("Scatter Plot of Noise (First 1000 Points)")
plt.xlabel("Index")
plt.ylabel("Noise Value")
plt.colorbar(label="Intensity")
plt.show()
```

Task 5 Random 2D data scatter [ 2 clasters, different sizes) 

```python
np.random.seed(42)

# Cluster 1: larger, more spread
n1 = 300
mean1 = [2, 3]
cov1 = [[1.5, 0], [0, 1.5]]
cluster1 = np.random.multivariate_normal(mean1, cov1, n1)

# Cluster 2: smaller, tighter
n2 = 100
mean2 = [-2, -1]
cov2 = [[0.4, 0], [0, 0.4]]
cluster2 = np.random.multivariate_normal(mean2, cov2, n2)

# Combine and plot
X = np.vstack([cluster1, cluster2]) #stacks the two (n2) arrays into one (4002) data matrix X.
labels = [0]*n1 + [1]*n2

plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='Set1', s=50, edgecolors='k', alpha=0.8)
plt.scatter(mean1[0], mean1[1], c='blue', s=200, marker='X', label=f'Center 1 (n={n1})')
plt.scatter(mean2[0], mean2[1], c='red', s=200, marker='X', label=f'Center 2 (n={n2})')
plt.title("Two Clusters with Different Sizes")
plt.xlabel("X1")
plt.ylabel("X2")
plt.legend()
plt.grid(True, alpha=0.3)
plt.axis('equal')
plt.show()
```
