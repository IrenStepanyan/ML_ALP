**Task 1: Visalize 3 functions**

```python
x = np.linspace(-np.pi,np.pi,50)
y_tgx = np.tan(x)
y_ctgx = 1/np.tan(x)
y_line = x**2
plt.plot(x,y_tgx)
plt.plot(x,y_ctgx)
plt.plot(x, y_line)
plt.legend(['tan(x)', 'ctan(x)', 'square of x']) #for the labeles 
plt.show()
```

**Task 2 2D normal distribution**

```python
mu = 100
sigma = 5
size = 100000
 
# creating a normal distribution data
values = np.random.normal(mu, sigma, size)
 
# plotting 
plt.hist(values, 100)

# plotting mean line
plt.axvline(values.mean(), color='red')
plt.show()
```
